### Fill Value:
##### Sometimes, the satellite can‚Äôt measure the real value (maybe it‚Äôs broken, or out of range).

##### Instead of leaving it blank, it puts a special number like -1e31 or 9999999 ‚Äî that number is called the "fill value".

##### This line fetches that special number from the metadata of the variable.

```python
meta = cdf.varattsget('proton_bulk_speed')
print("Fill value:", meta.get('FILLVAL'))
```

## Ploting proton bulk speed


```python
import matplotlib.pyplot as plt
from cdflib import CDF
from cdflib.epochs import CDFepoch
import numpy as np

# Load the file
file_path = r'C:\Users\Lenovo\Desktop\Swis_Project\swis_data\AL1_ASW91_L2_BLK_20250324_UNP_9999_999999_V02.cdf'
cdf = CDF(file_path)

# Convert time
times = CDFepoch.to_datetime(cdf['epoch_for_cdf_mod'][:])

# Get data
bulk_speed = cdf['proton_bulk_speed'][:].flatten()

# Get fill value from metadata
fill_val = cdf.varattsget('proton_bulk_speed').get('FILLVAL')

# Filter out fill values and other extreme outliers
valid_mask = (bulk_speed != fill_val) & (bulk_speed > 0) & (bulk_speed < 2000)

# Apply mask
times_clean = np.array(times)[valid_mask]
bulk_speed_clean = bulk_speed[valid_mask]

# Plot cleaned data
plt.figure(figsize=(12, 5))
plt.plot(times_clean, bulk_speed_clean, label='Proton Bulk Speed [km/s]', color='green')
plt.xlabel('Time (UTC)')
plt.ylabel('Speed (km/s)')
plt.title('Cleaned Proton Bulk Speed Over Time')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
import os
import numpy as np
import pandas as pd
from cdflib import CDF
from cdflib.epochs import CDFepoch

# === Folder with CDF files ===
data_folder = r'C:\Users\Lenovo\Desktop\Swis_Project\swis_data'

# === Variables to Extract ===
variables = [
    'epoch_for_cdf_mod',
    'proton_density',
    'proton_bulk_speed',
    'proton_thermal',
    'alpha_density'
]

# === Initialize storage ===
all_data = {
    'time': [],
    'proton_density': [],
    'proton_bulk_speed': [],
    'proton_thermal': [],
    'alpha_density': []
}

# === Loop through all CDF files ===
for filename in sorted(os.listdir(data_folder)):
    if filename.endswith('.cdf'):
        file_path = os.path.join(data_folder, filename)
        cdf = CDF(file_path)
        
        try:
            time = CDFepoch.to_datetime(cdf['epoch_for_cdf_mod'][:])
        except Exception as e:
            print(f"Skipping {filename}: {e}")
            continue

        all_data['time'].extend(time)

        for var in variables[1:]:  # skip time
            data = cdf[var][:].flatten()
            fill_val = cdf.varattsget(var).get('FILLVAL')
            # Replace fill values with NaN
            cleaned = np.where(data == fill_val, np.nan, data)
            all_data[var].extend(cleaned)

# === Convert to DataFrame ===
df = pd.DataFrame(all_data)
df['time'] = pd.to_datetime(df['time'])
df = df.set_index('time').sort_index()

# === (Optional) Save to CSV for reuse ===
# df.to_csv("merged_swis_data.csv")
```

### Ploting some variable to see the spike in Hello CME event

```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from cdflib import CDF
from cdflib.epochs import CDFepoch

# === Folder with all your CDF files ===
data_folder = r'C:\Users\Lenovo\Desktop\Swis_Project\swis_data'

# === Variables to Load ===
variables = ['epoch_for_cdf_mod', 'proton_density', 'proton_bulk_speed', 'proton_thermal', 'alpha_density']

# === Storage for Merged Data ===
all_data = {var: [] for var in variables}
timestamps = []

# === Load and Merge All Files ===
for filename in sorted(os.listdir(data_folder)):
    if filename.endswith('.cdf'):
        path = os.path.join(data_folder, filename)
        cdf = CDF(path)
        try:
            time = CDFepoch.to_datetime(cdf['epoch_for_cdf_mod'][:])
            timestamps.extend(time)

            for var in variables[1:]:  # skip epoch
                data = cdf[var][:].flatten()
                fill_val = cdf.varattsget(var).get('FILLVAL')
                data = np.where(data == fill_val, np.nan, data)
                all_data[var].extend(data)
        except Exception as e:
            print(f"‚ùå Skipped {filename} due to error: {e}")

# === Build DataFrame ===
df = pd.DataFrame({
    'time': pd.to_datetime(timestamps),
    'proton_density': all_data['proton_density'],
    'proton_bulk_speed': all_data['proton_bulk_speed'],
    'proton_thermal': all_data['proton_thermal'],
    'alpha_density': all_data['alpha_density'],
}).set_index('time').sort_index()

# === Drop rows with all NaNs ===
df = df.dropna(how='all')

# === Derived Features ===
density = df['proton_density']
speed = df['proton_bulk_speed']
thermal = df['proton_thermal']
alpha = df['alpha_density']

alpha_proton_ratio = (alpha / density).replace([np.inf, -np.inf], np.nan)
dyn_pressure = 1.6726e-6 * density * speed**2 * 1e-9  # in Pa

# === Moving Averages ===
ma_density_15m = density.rolling('15min').mean()
ma_density_3h = density.rolling('3h').mean()
ma_speed_15m = speed.rolling('15min').mean()
velocity_gradient = speed.diff().rolling('15min').mean() * 100  # scaled

# === Plot All in 5 Subplots ===
fig, axs = plt.subplots(5, 1, figsize=(14, 12), sharex=True)
fig.suptitle("SWIS Level-2 Data & Derived Features (Merged)", fontsize=14)

# Panel 1: Proton Density
axs[0].plot(density.index, density, label='Proton Density (Raw)', alpha=0.6)
axs[0].plot(ma_density_15m.index, ma_density_15m, label='Density MA (15min)', color='orange')
axs[0].plot(ma_density_3h.index, ma_density_3h, label='Density MA (3h)', color='brown', linestyle='dotted')
axs[0].set_ylabel("Density (cm‚Åª¬≥)")
axs[0].legend()
axs[0].grid(True)

# Panel 2: Proton Bulk Speed
axs[1].plot(speed.index, speed, label='Bulk Speed (Raw)', alpha=0.6)
axs[1].plot(ma_speed_15m.index, ma_speed_15m, label='Speed MA (15min)', color='goldenrod')
axs[1].plot(velocity_gradient.index, velocity_gradient, label='Velocity Gradient (x100)', linestyle='--', color='red')
axs[1].set_ylabel("Speed (km/s)")
axs[1].legend()
axs[1].grid(True)

# Panel 3: Thermal Speed
axs[2].plot(thermal.index, thermal, label='Proton Thermal Speed', color='green')
axs[2].set_ylabel("Thermal Speed (km/s)")
axs[2].legend()
axs[2].grid(True)

# Panel 4: Alpha-to-Proton Ratio
axs[3].plot(alpha_proton_ratio.index, alpha_proton_ratio, color='purple', label='Alpha/Proton Ratio')
axs[3].set_ylabel("Alpha/Proton Ratio")
axs[3].legend()
axs[3].grid(True)

# Panel 5: Dynamic Pressure
axs[4].plot(dyn_pressure.index, dyn_pressure, color='blue', label='Dynamic Pressure (Pa)')
axs[4].set_ylabel("Dynamic Pressure (Pa)")
axs[4].legend()
axs[4].grid(True)

# Final Formatting
axs[4].set_xlabel("Time (UTC)")
plt.tight_layout()
plt.subplots_adjust(top=0.93)
plt.show()
```
